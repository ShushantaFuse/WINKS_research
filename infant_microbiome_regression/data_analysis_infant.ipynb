```python
cols = otu_data.columns
otu_data[cols] = otu_data[cols].apply(pd.to_numeric, errors='coerce')
```

```python
otu_data["sum"] = otu_data.sum(axis=1)
```

```python
otu_data.loc[:, otu_data.columns != '#OTU ID']=otu_data.loc[:, otu_data.columns != '#OTU ID'].div(otu_data["sum"], axis=0)
otu_data
```

```python
merge_data= pd.merge(baby_df,otu_data,left_on=['#SampleID'],right_index=True)
```

```python
Y = merge_data["Var"]
merge_data.drop(["#SampleID","Var","sum"],inplace=True,axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(merge_data, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

from sklearn.metrics import mean_absolute_error,mean_squared_error
 
mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
#squared True returns MSE value, False returns RMSE value.
mse = mean_squared_error(y_true=y_test,y_pred=y_pred) #default=True
rmse = mean_squared_error(y_true=y_test,y_pred=y_pred,squared=False)
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```
