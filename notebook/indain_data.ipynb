```python
id=df['Subject_ID']
y=df['BMI']
y
```

```python
cols=['Subject_ID', 'Geographical Location',
       'Gender', 'AGE  in years', 'Life style pattern']
df=df.loc[:,cols]
```

```python
import scipy
from scipy import stats
df['AGE  in years']=stats.zscore(df['AGE  in years'])
```

```python
df.head()
```

```python
one_hot_encoded_data = pd.get_dummies(df,columns = ['Geographical Location', 'Gender','AGE  in years','Life style pattern'])
one_hot_encoded_data
```

```python
filename=os.path.join(dir,"indian_gut_microbiota_data\\pan_otutab.csv")
otu = pd.read_csv(filename).T
print(otu.head())
```

```python
otu
```

```python
otu.columns = otu.iloc[0] 

otu = otu[1:]
otu
```

```python
cols = otu.columns
otu[cols] = otu[cols].apply(pd.to_numeric, errors='coerce')

otu=otu.loc[:, (otu[otu > 0].count()>= 5)] 
otu

```

```python
otu=otu.reset_index()
otu.index
```

```python
otu
```

```python
merge_data= pd.merge(one_hot_encoded_data,otu,left_on=['Subject_ID'],right_on=['index'])
merge_data
```

```python
merge_data=merge_data.drop(['index', 'Subject_ID'], axis = 1)
merge_data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(merge_data, y, test_size=0.2, random_state=42)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

from sklearn.metrics import mean_absolute_error,mean_squared_error
 
mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
#squared True returns MSE value, False returns RMSE value.
mse = mean_squared_error(y_true=y_test,y_pred=y_pred) #default=True
rmse = mean_squared_error(y_true=y_test,y_pred=y_pred,squared=False)
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
from sklearn.svm import SVR
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3)
 
# fitting the model for grid search
grid.fit(X_train, y_train)

```

```python
grid.best_params_
```

```python
regressor=SVR(C= 1000, gamma= 0.0001,kernel='rbf')
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
#squared True returns MSE value, False returns RMSE value.
mse = mean_squared_error(y_true=y_test,y_pred=y_pred) #default=True
rmse = mean_squared_error(y_true=y_test,y_pred=y_pred,squared=False)
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```
