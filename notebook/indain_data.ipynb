```python
id=df['Subject_ID']
y=df['BMI']
y
```

```python
cols=['Subject_ID', 'Geographical Location',
       'Gender', 'AGE  in years', 'Life style pattern']
df=df.loc[:,cols]
```

```python
df.head()
```

```python
one_hot_encoded_data = pd.get_dummies(df,columns = ['Geographical Location', 'Gender','AGE  in years','Life style pattern'])
one_hot_encoded_data
```

```python
filename=os.path.join(dir,"indian_gut_microbiota_data\\pan_otutab.csv")
otu = pd.read_csv(filename).T
print(otu.head())
```

```python
otu
```

```python
otu.columns = otu.iloc[0] 

otu = otu[1:]
otu
```

```python
cols = otu.columns
otu[cols] = otu[cols].apply(pd.to_numeric, errors='coerce')

otu=otu.loc[:, (otu[otu > 0].count()>= 5)] 
otu

```

```python
otu=otu.reset_index()
otu.index
```

```python
otu
```

```python
merge_data= pd.merge(one_hot_encoded_data,otu,left_on=['Subject_ID'],right_on=['index'])
merge_data
```

```python
merge_data=merge_data.drop(['index', 'Subject_ID'], axis = 1)
merge_data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(merge_data, y, test_size=0.2, random_state=42)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

from sklearn.metrics import mean_absolute_error,mean_squared_error
 
mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
#squared True returns MSE value, False returns RMSE value.
mse = mean_squared_error(y_true=y_test,y_pred=y_pred) #default=True
rmse = mean_squared_error(y_true=y_test,y_pred=y_pred,squared=False)
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
from sklearn.svm import SVR
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3)
 
# fitting the model for grid search
grid.fit(X_train, y_train)

```

```python
grid.best_params_
```

```python
regressor=SVR(C= 1000, gamma= 0.0001,kernel='rbf')
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
#squared True returns MSE value, False returns RMSE value.
mse = mean_squared_error(y_true=y_test,y_pred=y_pred) #default=True
rmse = mean_squared_error(y_true=y_test,y_pred=y_pred,squared=False)
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```

```python
from sklearn.ensemble import RandomForestRegressor
param_grid = { 
    'n_estimators': [100, 200,250,300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [2,3,4,5],
    'criterion' :['squared_error', 'absolute_error', 'poisson']
}
rfr=RandomForestRegressor(random_state=42)
```

```python
CV_rfc = GridSearchCV(estimator=rfr, param_grid=param_grid)
CV_rfc.fit(X_train, y_train)
```

```python
CV_rfc.best_params_
```

```python
from sklearn.inspection import permutation_importance
perm_importance = permutation_importance(regressor, X_test, y_test)
features = np.array(merge_data.columns)
sorted_idx = perm_importance.importances_mean.argsort()[0:5]
plt.barh(features[sorted_idx], perm_importance.importances_mean[sorted_idx])
plt.xlabel("Permutation Importance")
```

```python
filename=os.path.join(dir,"indian_gut_microbiota_data\\pan_meta.csv")
df = pd.read_csv(filename)
sns.countplot(data=df,x='Obese/Non Obese ')
y=df['Obese/Non Obese ']
```

```python
X_train, X_test, y_train, y_test = train_test_split(merge_data, y, test_size=0.2, random_state=42)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.model_selection import GridSearchCV
param_grid = { 
    'n_estimators': [100, 200,250,300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [2,3,4,5],
    'criterion' :['gini', 'entropy']
}
rfc=RandomForestClassifier(random_state=42)
```

```python
y
```

```python
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 3)
CV_rfc.fit(X_train, y_train)
CV_rfc.best_params_
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
rfc1=RandomForestClassifier(random_state=42, max_features='auto', n_estimators= 100, max_depth=4, criterion='gini')
rfc1.fit(X_train,y_train)
y_pred=rfc1.predict(X_test)
print("Accuracy for Random Forest on test data: ",accuracy_score(y_test,y_pred))
```

```python
y_pred_proba=rfc1.predict_proba(X_test)
print(" ROC score for Random Forest on test data: ",roc_auc_score(y_test,y_pred_proba[:,1]))
```
