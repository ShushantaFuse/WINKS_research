```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

from pathlib import Path
path=Path(os.getcwd())
dir=path.parent.absolute()
```

```python
dir=os.path.join(dir,"Duvallet")
files=os.listdir(dir)
```

```python
df = pd.DataFrame()
for i in range(0, len(files),2):
    filename=os.path.join(dir,files[i+1])
    meta = pd.read_csv(filename,encoding='latin1')
    print(meta.shape)
    meta=meta[['Unnamed: 0','DiseaseState']]
    filename=os.path.join(dir,files[i])
    otu=pd.read_csv(filename)
    print(otu.shape)
    data= pd.merge(meta,otu,on='Unnamed: 0')
    if(i==0):
        df=data
    else:
        df=df.append([data])

```

```python
df["DiseaseState"] = np.where(df["DiseaseState"] == "H", 1, 0)
```

```python
y=df['DiseaseState']
ids=df['Unnamed: 0']
df=df.drop(['DiseaseState','Unnamed: 0'], axis = 1)
df
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
df.replace(np.nan, 0,inplace=True)
df.replace([np.inf, -np.inf], 0, inplace=True)
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42,stratify=y)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.model_selection import GridSearchCV
param_grid = { 
    'n_estimators': [100, 200,250,300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [3,4,5,6,7],
    'criterion' :['gini', 'entropy']
}
rfc=RandomForestClassifier(random_state=42)
```

```python
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 7)
CV_rfc.fit(X_train, y_train)
CV_rfc.best_params_
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
rfc1=RandomForestClassifier(random_state=42, max_features='auto', n_estimators= 200, max_depth=7, criterion='gini')
rfc1.fit(X_train,y_train)
y_pred=rfc1.predict(X_test)
print("Accuracy for Random Forest on test data: ",accuracy_score(y_test,y_pred))
```

```python
y_pred_proba=rfc1.predict_proba(X_test)
print(" ROC score for Random Forest on test data: ",roc_auc_score(y_test,y_pred_proba[:,1]))
```
