```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

from pathlib import Path
path=Path(os.getcwd())
dir=path.parent.absolute()
```

```python
# data link: https://ibdmdb.org/tunnel/public/summary.html
filename=os.path.join(dir,"diet_data\\task.txt")
df = pd.read_csv(filename, sep ='\t')
print(df.head)
```

```python
df
```

```python
filename=os.path.join(dir,"diet_data\\taxatable.txt")
otu=pd.read_csv(filename, sep="\t").T
otu
```

```python
otu[:30]
```

```python
otu.columns = otu.iloc[0] 

otu = otu[1:]
otu
```

```python
cols = otu.columns
otu[cols] = otu[cols].apply(pd.to_numeric, errors='coerce')
```

```python
otu["sum"] = otu.sum(axis=1)
otu
```

```python
otu.loc[:, otu.columns != '#OTU ID']=otu.loc[:, otu.columns != '#OTU ID'].div(otu["sum"], axis=0)
otu
```

```python
df.index
```

```python
data= pd.merge(df,otu,left_on='#SampleID', right_index=True)
```

```python
data
```

```python
y=data['Var']
data=data.drop(['Var', '#SampleID','sum'], axis = 1)
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(data)
from sklearn import decomposition
pca = decomposition.PCA()
pca.n_components = 10
sample_data=standardized_data
pca_data = pca.fit_transform(sample_data)
# Plot the PCA spectrum
percentage_var_explained = pca.explained_variance_ / np.sum(pca.explained_variance_);
cum_var_explained = np.cumsum(percentage_var_explained)
plt.figure(1, figsize=(6, 4))
plt.clf()
plt.plot(cum_var_explained, linewidth=2)
plt.axis('tight')
plt.grid()
plt.xlabel('n_components')
plt.ylabel('Cumulative_explained_variance')
plt.show()
```

```python
pca.n_components = 2
pca_data = pca.fit_transform(sample_data)
pca_data = np.vstack((pca_data.T, y)).T
# creating a new data fram which help us in ploting the result data
pca_df = pd.DataFrame(data=pca_data, columns=("1st_principal", "2nd_principal", "status"))
sns.FacetGrid(pca_df, hue="status", size=6).map(plt.scatter, '1st_principal', '2nd_principal').add_legend()
plt.show()
```

```python
y.replace({'LowFat': 0, 'HighFat': 1}, inplace=True)
y.value_counts()
```

```python
from mpl_toolkits.mplot3d import Axes3D
pca.n_components = 3
pca_data = pca.fit_transform(sample_data)
print("shape of pca_reduced.shape = ", pca_data.shape)
pca_data = np.vstack((pca_data.T, y)).T
pca_df = pd.DataFrame(data=pca_data, columns=("1st_principal", "2nd_principal", "3rd_principal","status"))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
c_map = {0: 'blue', 1: 'red'} 
ax.scatter(pca_df['1st_principal'], pca_df['2nd_principal'], pca_df['3rd_principal'], c=[c_map[_] for _ in y], s=20)
ax.view_init(30, 185)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.2, random_state=42,stratify=y)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
params = {'max_leaf_nodes': list(range(2, 100)), 'min_samples_split': [2, 3, 4],'criterion' :['gini', 'entropy']}

grid_search_cv = GridSearchCV(DecisionTreeClassifier(random_state=42), params, verbose=1, cv=3)
```

```python
grid_search_cv.fit(X_train, y_train)
grid_search_cv.best_params_
```

```python
clf=DecisionTreeClassifier(max_leaf_nodes=2, random_state=42,min_samples_split= 2)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
y_pred=clf.predict(X_test)
print("Accuracy for Decision tree on test data: ",accuracy_score(y_test,y_pred))
y_pred_proba=clf.predict_proba(X_test)
print("AUCROC score for Decision tree on test data:",roc_auc_score(y_test,y_pred_proba[:,1]))
```

```python
from sklearn.svm import SVC
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
grid = GridSearchCV(SVC(), param_grid, cv=3, verbose = 3)

grid.fit(X_train, y_train)
```

```python
y_pred=grid.predict(X_test)
print("Accuracy for Decision tree on test data: ",accuracy_score(y_test,y_pred))
y_pred_proba=grid.predict_proba(X_test)
print("AUCROC score for Decision tree on test data:",roc_auc_score(y_test,y_pred_proba[:,1]))
```
